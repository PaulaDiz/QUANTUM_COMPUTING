```python
from qiskit import QuantumCircuit
from qiskit import QuantumRegister, ClassicalRegister
#CREACIÓN DEL CIRCUITO

#Número de qubits
qubit = QuantumRegister(4, 'q') #qubits cuanticos

#Número de bits (Cada qubit lleva un bit asociado)
bit = ClassicalRegister(4, 'our_bit') #bits clasicos

#Creamos el circuito con nuestros qubits y bits
qc = QuantumCircuit(qubit, bit) # con (1,1) tambien funciona primero qubits segundo bits clasicos
#-------------------------------------------------------------------------------------------------

#1.SUPERPOSICIÓN INICIAL
qc.x(3)

qc.h(0) #qubit[0]seria si tuvieramos dos y lo aplicariamos al segundo
qc.h(1)
qc.h(2)
qc.h(3)
#--------------------------------------------------------------------------------------------------
#ORÁCULO
qc.barrier()

qc.cx(0,3) # q0 es el cubit de control y q1 el que actua
qc.cx(1,3)
qc.cx(2,3)
#---------------------------------------------------------------------------------------------------
qc.barrier()

qc.h(0) #qubit[0]seria si tuvieramos dos y lo aplicariamos al segundo
qc.h(1)
qc.h(2)
qc.h(3)
#--------------------------------------------------------------------------------------------------
#MEDICIONES
qc.measure(0,0)
qc.measure(1,1)
qc.measure(2,2)
qc.measure(3,3)
#tambien puede hacerse como -> qc.measure([0,0],[1,1]) el primer corchete son los cubits y el segundo
#los bits asociados a cada uno.

#----------------------------------------------------------------------------------------------------
#DIBUJUAMOS EL CIRCUITO
qc.draw(output='mpl')
```
